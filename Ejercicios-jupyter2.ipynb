## 2.1

Para graficar un numero complejo y su conjugado en el plano complejo, uso la funcion definida previamente en los ejemplos dados y la modifico convenientemente para poder graficar. 

```python
import cmath
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display

x = complex(3,4)

def plot(x):
    x_conjugate = x.conjugate()

    fig, ax = plt.subplots()
    ax.plot([0, x.real], [0, x.imag], 'ro-')
    ax.plot([0, x_conjugate.real], [0, x_conjugate.imag], 'bo-')
    ax.set_xlim(-5, 5)
    ax.set_ylim(-5, 5)
    ax.set_xlabel('Real')
    ax.set_ylabel('Imaginary')
    ax.axhline(0, color='black',linewidth=0.5)
    ax.axvline(0, color='black',linewidth=0.5)
    ax.grid(True)
    plt.title("Numero complejo y conjugado")
    plt.show()

style = {'description_width': 'initial'}
layout = widgets.Layout(width='400px')


plot(x)

```

## Exercise 3: Exploring the Mandelbrot Set
1. Modify the provided Mandelbrot set code to change its zoom level and center point. Observe how the fractal pattern changes.
2. Experiment with different values of `max_iter` and observe the effect on the fractal's detail and computation time.

## 3.1


```python
# Example: Generating a Mandelbrot set

import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_iter):
    z = 0
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iter

# Create a grid of complex numbers
pixels = 800
max_iter = 30
x = np.linspace(-1, 1, pixels)
y = np.linspace(-2, 2, pixels)
xx, yy = np.meshgrid(x, y)
c = xx + yy*1j

# Compute Mandelbrot set
mandelbrot_set = np.array([mandelbrot(ci, max_iter) for ci in c.ravel()]).reshape(c.shape)

# Plot
plt.imshow(mandelbrot_set, extent=[-2, 1, -1.5, 1.5])
plt.colorbar()
plt.title("Mandelbrot Set")
plt.show()
```

Cuando hacemos zoom podemos ver como se pierden un poco los patrones que tiene el fractal, pero cuando alejamos la imagen los patrones se hacen mas evidentes como se puede evidenciar en la segunda imagen.

## Exercise 4: Creating a Julia Set
1. Implement a Julia set generator. Use a constant like `-0.4 + 0.6j` for the iteration `z = z*z + constant`.
2. Explore how changing the constant changes the pattern of the Julia set.

## 4.1


```python
import numpy as np
import matplotlib.pyplot as plt

c = 0.37 + 0.16j

def julia(c, max_iter, z):
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iter

# Create a grid of complex numbers
pixels = 800
max_iter = 30
x = np.linspace(-2, 2, pixels)
y = np.linspace(-2, 2, pixels)
xx, yy = np.meshgrid(x, y)
z = xx + yy*1j

# Compute julia set
julia_set = np.array([julia(c, max_iter, zi) for zi in z.ravel()]).reshape(z.shape)

# Plot
plt.imshow(julia_set, extent=[-2, 1, -1.5, 1.5])
plt.colorbar()
plt.title("Julia Set")
plt.show()
```

Al implementar el conjunto de Julia con los valores indicados, podemos ver como emergen patrones unicos. Aunque el calculo es similar al del conjunto Mandelbrot, la principal diferncia es que usamos una constante especifica en la ecuacion. Esta constante es la que nos define cuales son los patrones que va a tener el fractal.
